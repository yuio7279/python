# 9. 데이터 선택 (조건)

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df['키'] >= 185 # 학생들의 키가 185 이상인지를 bool
```

```python
filt = (df['키'] >= 185) #filter변수 설정
df[filt]
```

```python
df[~filt] #filter를 역으로 적용
```

```python
df[df['키']>=185] #변수 없이 바로 설정
```

```python
df.loc[df['키']>=185, '수학'] #키 185이상의 수학점수
```

```python
df.loc[df['키']>=185,['이름','수학','과학']] #키가 185이상의 이름 수학 과학 데이터
```

## 다양한 조건

### & 그리고

```python
df.loc[(df['키']>=185) & (df['학교'] =='북산고')] #키가 185 이상의 북산고 학생  데이터
```

### |또는

```python
df.loc[(df['키']<170 )|(df['키']>200)] #170보다작거나 200보다 크거나
```

### str함수

```python
filt = df['이름'].str.startswith('송') #송 씨 성을 가진사람
```

```python
df[filt]
```

```python
filt = df['이름'].str.contains('태') #이름에 '태' 가 들어가는 사람
df[filt]
```

```python
df[~filt] #이름에 '태'가 들어가는 사람을 제외
```

```python
langs = ['Python','Java']
filt = df['SW특기'].isin(langs) #SW특기가 Python 이거나 Java인 사람
df[filt]
```

```python
df #대문자 PYTHON을 가져오지 못했다
```

```python
langs =  ['python','java']
filt = df['SW특기'].str.lower().isin(langs)
df[filt]
```

```python
filt = df['SW특기'].str.contains('Java')
df[filt] #NaN처리하지 못해 에러
```
